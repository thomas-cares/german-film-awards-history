```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
url='https://www.deutscher-filmpreis.de/historie/page/1/'
### https://www.deutscher-filmpreis.de/historie/page/105/ scheint letzte Seite zu sein 1951
pages=105
rows=10
arranger=[i*5 for i in range(rows)]#helps to arrange the information in the rows
```

```python
def make_soup(url):
    html=requests.get(url).content
    return BeautifulSoup(html,'lxml')

def list_creator(pages):
    lst=[]
    for i in range(pages):
        url=f'https://www.deutscher-filmpreis.de/historie/page/{i}/'
        soup=make_soup(url)
        for x in arranger:
            lst.append([i.text.replace('\nKategorie\n','').replace('\nFilmtitel\n','').replace('\nPreisträger\n','').replace('\nPreis\n','').replace('Jahr\n','').strip() for i in soup.find_all('td',{'class':'table-cell'})][x:x+5])
    return lst
```

```python
lst=list_creator(pages)
```

```python
df=pd.DataFrame(lst)
df.columns = ['Jahr','Kategorie','Filmtitel','PreisträgerInnen','Preisname']
df
```

## EXPORT

```python
df.to_csv('DFP_Historie.csv')
```
